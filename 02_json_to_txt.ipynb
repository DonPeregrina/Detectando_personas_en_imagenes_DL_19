```python
import json
from pprint import pprint

folder = 'dataset_intrusos/'  
with open(folder + 'dataset_intrusos.json') as f:  
    data = json.load(f)

pprint(data) #Visualizar estructura de datos
```

Row format: image_file_path box1 box2 ... boxN;

Box format: x_min,y_min,x_max,y_max,class_id (no space).

Here is an example:

path/to/img1.jpg 50,100,150,200,0 30,50,200,120,3

path/to/img2.jpg 120,300,250,600,2

...


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

clase_id = 0
file = open('dataset_intrusos.txt','w') 
true_detections = {}
for item in data['_via_img_metadata'].items(): 
    filename = item[1]['filename']
    string = folder + filename
    for region in item[1]['regions']:
        x = region['shape_attributes']['x']
        y = region['shape_attributes']['y']
        h = region['shape_attributes']['height']
        w = region['shape_attributes']['width']
        string = string + ' ' + str(x) + ','  + str(y) + ','  + str(x+w) + ',' + str(y+h) + ',' + str(clase_id)
    print(string)
    file.write(string)
    file.write('\n') 
    
file.close()
```
